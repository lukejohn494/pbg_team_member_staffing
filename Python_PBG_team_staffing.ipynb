```python
#Penn Biotech Group
#Staffing algorithm Python script
#Written by VP Knowledge Management - Luke Johnson (2018-2019)
#Adapted from previous staffing algorithm written in R

#Read in applicant data and a control file

#applicant_data should have the following header row:
#c("name", "penn_email", "preference_1", "preference_2", "preference_3", "preference_4", "manual_staffing", "other_projects", "unlisted_projects")

#control_file should have the following header row:
#c("team_name", "team_size")

#The team names from the applicant data file as listed under preference_1, etc. and the control file team names must be EXACTLY THE SAME

#Likewise, the team name listed for any manual staffing must be exactly the same

#Other projects and unlisted projects should have yes/no answers

#Ensure that none of the fields within either the applicant data or control file contains commas

#These are COMMA SEPARATED files, so any teams names listed as "TEAM X, LLC" will mess up reading in the file.

#CHANGE THESE TWO LINES TO THE PROPER PATH FOR APPLICANT DATA AND CONTROL FILE

import numpy

## 3 or 4 times, save the file

from datetime import datetime

# this was done to attach the data of creation of an output file

d=datetime.today()
d1=str(d.month)+str(d.day)+str(d.year)
d2=str(d.hour)
d3=str(d.minute)
d4=str(d.second)



applicant_data=numpy.loadtxt("staffing_input_applicant_data.csv", delimiter=',', dtype=numpy.str)

control_file=numpy.loadtxt("staffing_input_control_file.csv", delimiter=',', dtype=numpy.str)

#head(applicant_data)
#head(control_file)

print(applicant_data)
print(control_file)
print(applicant_data.shape)
print(control_file.shape)
print(float('5'))
print(control_file[1,1])
```

```python
#Create an empty staffing matrix to fill. Dimensions: # of teams rows and maximum team size (of any team) # columns
print([int(float(control_file[i,1])) for i in range(1, control_file.shape[0])])
print(max([int(float(control_file[i,1])) for i in range(1, control_file.shape[0])]))

#For each of the teams that will be staffed with fewer than the maximum team size, add BUFFER text 

empty_staffing=numpy.chararray((control_file.shape[0]-1,max([int(float(control_file[i,1])) for i in range(1, control_file.shape[0])])), itemsize=41)
empty_staffing[:]='BUFFERMEOHBUFFERIEVERYONELOVESAGOODBUFFER'
empty_staffing=empty_staffing.decode("utf-8")
print(empty_staffing)
print(empty_staffing.shape[0])
#empty_staffing
```

```python
#Fill in with no 'BUFFERMEOHBUFFERIEVERYONELOVESAGOODBUFFER'
#Needs to correspond with group size in control size file

for y in range(0,empty_staffing.shape[0]):
    z=int(control_file[y+1,1])
    while z<empty_staffing.shape[1]:
        empty_staffing[y,z]='NOTGOODBUFFATALL'
        z+=1
#Fill in empty_staffing with manual overrides first
counter=[int(0) for i in range(0, control_file.shape[0])]

for i in range(1,applicant_data.shape[0]):
    if applicant_data[i,6]!='':
        for count, value in enumerate(control_file[1:,0]):
            if value==applicant_data[i,6]:
                empty_staffing[count,counter[count]]=applicant_data[i,0]
                counter[count]+=1
            
print(empty_staffing)
#empty_staffing
```

```python
#Remove manually staffed applicants from the applicant data

from copy import deepcopy

ad1=deepcopy(applicant_data) #made a copy of applicant_data to shorten variable
print(numpy.sum(empty_staffing != "BUFFERMEOHBUFFERIEVERYONELOVESAGOODBUFFER"))

print(applicant_data.shape)

#Shuffle remaining applicants

k=0
county=0
collect=[]
for k in range(0,applicant_data.shape[0]):
    if ad1[k,0] in empty_staffing:
        collect.append(k)

ad2=deepcopy(applicant_data)
ctup=tuple(collect)
##better way to remove the rows from the numpy file
ad2=numpy.delete(ad2,ctup,0)
print(ad2)
```

```python
#Shuffle the rows

ad3=ad2[1:,:]
numpy.random.shuffle(ad3)
ad2=numpy.vstack((ad2[0,:],ad3))
#print(ad2)

import random
my_list = ['2'] * 70 + ['3'] * 30

ad33=ad2[0,:]

count4=0
for jj in range(0,ad2.shape[0]):
    if ad2[jj,9]=='4':
        count4+=1 

print(count4)

ci=1
while ad33.shape[0]<ad2.shape[0]-count4:
    vr=random.choice(my_list)
    #print(ad2[ci,0] not in ad33)
    if ad2[ci][9]==vr and ad2[ci,0] not in ad33:
        ad33=numpy.vstack((ad33,ad2[ci,:]))
    ci+=1
    if ci==ad2.shape[0]:
        ci=1
    #print(ad33.shape[0], ad2.shape[0]-count4)
    #print(ad33)
    #print(ci)
#print(ad33)

for jj in range(0,ad2.shape[0]):
    if ad2[jj,9]=='4':
        ad33=numpy.vstack((ad33,ad2[jj,:]))

ad2=ad33
print(ad2)
```
